```python
########## Setting up Datasets ##########
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import timeit
import seaborn as sns; sns.set()

training_df= pd.read_csv('training.txt')
test_df = pd.read_csv('test.txt')
```

```python
sns.set(rc={'figure.figsize':(10, 10)})
plt.scatter(training_df.x, training_df.y, c = training_df.colour, s =100)
plt.title('Training Dataset')
plt.ylabel('y')
plt.xlabel('x')
```

```python
# Setup Train/Test datasets
X_train = training_df.iloc[:,0:2].values
y_train = training_df.iloc[:, 2].values
X_test = test_df.iloc[:, 0:2].values
y_test = test_df.iloc[:, 2].values
```

```python
# Encoding the colour of points
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_y = LabelEncoder()
# Encoding training dataset
y_train[0:] = labelencoder_y.fit_transform(y_train[0:])
# Encoding test dataset
y_test[0:] = labelencoder_y.transform(y_test[0:])
```

```python
# Feature Scaling
# This improves the speed of training
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
########## Creating NN ##########
# Start Timer
start_time = timeit.default_timer()
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initialising the NN
model = Sequential() # classifier is just the name of nn

# Adding the input layer and the first hidden layer
# uniform is for the weights, all initialised close to 0
model.add(Dense(10, kernel_initializer = 'uniform', activation = 'relu', input_dim = 2))

# Adding the second hidden layer
model.add(Dense(10, kernel_initializer = 'uniform', activation = 'tanh'))

# Adding the third hidden layer
#model.add(Dense(10, kernel_initializer = 'uniform', activation = 'tanh'))


# Adding the output layer
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the NN
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the model
model.fit(X_train, y_train, batch_size = 30, validation_split = 0.33, epochs = 10000, verbose=1)

# End Timer
elapsed = timeit.default_timer() - start_time
print("Time ELapsed: " + str(elapsed))
```

```python
########## Evaluating the Model ##########
# Predicting the Test set results with Model (Not History class)
y_pred = model.predict(X_test).round()
# Convert u_test to correct dimmensional array for CM
y_test = np.expand_dims(y_test, axis=1)
y_test = y_test.astype('float32')
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy_val = accuracy_score(y_test,y_pred)
```

```python
# Visual of the NN
from keras.utils import plot_model
#plot_model(model, show_shapes = True, expand_nested = True)
#model.summary()

plt.figure(figsize=(12,12))

# Visual of Test Dataset values
plt.subplot(2,2,2, aspect = 'equal', adjustable = 'box')
plt.tight_layout()
plt.scatter(test_df.x, test_df.y, c = test_df.colour, s =100)
plt.title('Test Set')
plt.ylabel('y')
plt.xlabel('x')

# Visual of predicted values with Test Dataset
X_df = pd.DataFrame(X_test, columns = ['x','y'])
y_df = pd.DataFrame(y_pred, columns = ['colour'])
accuracy_df = X_df.join(y_df, how = 'outer')
plt.subplot(2,2,1,aspect = 'equal', adjustable = 'box')
plt.tight_layout()
plt.scatter(accuracy_df.x, accuracy_df.y, c = accuracy_df.colour, s = 100)
plt.title('Test Set with Predicted Colour Values')
plt.ylabel('y')
plt.xlabel('x')

# Creating History Object from the trained model
history = model.history
# Only if fitting with keras history
plt.subplot(2,2,3)
plt.tight_layout()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy Whilst Training')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='lower right')

# Visual of Training Dataset value
plt.subplot(2,2,4, aspect = 'equal', adjustable = 'box')
plt.tight_layout()
plt.scatter(training_df.x, training_df.y, c = training_df.colour, s = 100)
plt.title('Training Set')
plt.ylabel('y')
plt.xlabel('x')

plt.show()
print(" ")
print(" ")
print("Confusion Matrix")
print(cm)
print(" ")
print("MA on Static Test Set: " + str(accuracy_val))
print(" ")
print("Time ELapsed: " + str(elapsed))
print(" ")
print(" ")
```
